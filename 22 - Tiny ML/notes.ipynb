Machine Learning solutions (called **models**) are stochastic in nature: they are not always guaranteed to be correct. Example of applications of Machien Learning are **Image classification**, able to tell whether a picture has got a cat or a dog; **Object detection**, able to precisely localize the position of an object in an image; **Semantic segmentation**, heavily used in autonomous cars and able to colorizing an entire image into different categories, where all the pixels for a certain thing are of the same color (Where are the lanes of the road? Where are the shoulders in the road?). In the embedded system design there is also some issues in the software development side. Typically, what you find a three levels of abstractions: the high-level application, the libraries that provide support for those applications and then there's the operating system that provides support for the libraries and the applications.

<img src="./images/14.png" width="500">

The critical difference is on the operating system level. On a standard computing system, the operating system (Windows, Mac, or Linux for laptop, desktop and servers, or Apple iOS and Android for smartphones) provides a lot of flexibility and on top of that, developer can build different kinds of applications. When you go to an embedded system there isn't a similar support from an operating system, because typically the embedded system is really specialized to perform a single task. Some example of operating systems exist (FreeRTOS and Arm Mbed OS), however typically we don't tend to have these installed, because when we add any kind of extra stack on the embedded system, it takes away resources (and we are already constrained).

Also at the level of libraries we don't have uniform and standard situation. For example, in a standard system, libraries like Python NumPy are almost ubiquitously
available on any system: once you write the code using it,  it can run on virtually any system (Intel or AMD processor does not matter) and uniformly. You just focus on running your code. The library creates a level of abstraction and you don't need to worry about the nitty gritty details of the lower level. When we are considering an embedded system, this becomes a real problem. There is a trade-off there if we want universal code portability across any system, we pay in terms of cost and engineering effort. In an embedding system we forget portability and we want to be lean and efficient.

Consider AlexNet (2012), it predicts a thousand classes from ImageNet dataset with 57.1% of accuracy using a model of 61MB in size. Then as smartphones became prevalent for machine learning deployments, we saw the evolution move towards mobile nets. With mobile nets the size becomes extremely critical because models have to fit into the smartphone memory, so we start compromising the accuracy to dramatically shrunk the size of the network. Since then there have been other networks, but the critical point is that we understood the resource constraints of computing systems and  that we cannot just keep making things bigger. Bigger's not always better. However, the models that we have today are still pretty big. Our little embedded microcontrollers only have a few kilobytes of memory, it's an order of magnitude difference between where usually state of the art sits and how we need to cram things in.

**From a model perspective, we have to understand how we can you shrink or compress the model down without losing its ability to look for patterns in data?**

**Pruning** is the idea to take some of the connections of the network away, withouth degradating too much the performance of the ntwork. We might remove certain neurons completely, further reducing the amount of computation that's actually needed and shrinking the size of the network.

<img src="./images/21.png" width="500">

Another thing that we can do is play around with the numerical representation. Embedded systems don't have fancy hardware or they tend to do very simple sort of arithmetic, because they are supposed to be lean and mean as compared to general-purpose computing systems. **Quantization** means discretizing the floating point values down to a small subset of integer values. A floating value takes four bytes, an integer vakues is simply one byte, which means we get a 4x reductions in size. Moreover, floating point operations are complex and needs dedicated hardware, often not available on the embedded device, instead integer arithmetic is much more simpler and it is present on every device.

**Knowledge distillation** is tecnique in which a "teacher" (a big model), who is really good in performing a task, is able to distill down the critical information for a "student" (a small model), without losing
the ability to do the particular task.

The development of a tiny machine learning model follows a multi-step design methodology, commonly referred to as **lifecycle**, some of the most important steps are as follows:

<img src="./images/27.png" width="800">

The first stage is **design requirements**, where the application requirements are established. Hardware constraintsare often the most pressing requirements. Other requirements can include temporal resolution of data, the number of inferences per second, or minimum model accuracy. This step is the most crucial step, as it sets the scene. Then in **data Collection/Curation** we collect data that are suspected to aid in predicting the variable of interest. This can often involve active data collection from external sources, such as pre-existing datasets. This is often one of the most time consuming stages, as it can be difficult and time-consuming to curate a sufficiently large
dataset, simultaneously ensuring to minimize dataset bias. **Exploratory data analysis** analyzes the collect data to determine which features are most informative at predicting the variable of interest. Feature engineering is common at this stage, extracting new information from available data. **Preprocessing** involves ensuring the data follows standard model assumptions (e.g., are normally distributed) or manipulating data to meet these assumptions (e.g., log transformations on highly skewed variables). **Data imputation** is also typically done to appropriately fill in or remove missing data, and invalid or duplicate data is handled accordingly. All of the previous stages focused on the design aspects, as well as the procurement and analysis of data. These stages often make up the dominant proportion of the time in the machine learning lifecycle. Once a viable dataset is available, the next stage is **model development**: creating a suitable machine learning model. There are a plethora of machine learning techniques that have various pros and cons. One of the jobs of a machine learning practitioner is selecting an appropriate model for the task at hand. For example, basic linear regression may be suitable in environments where interpretability is key, but are constrained to the set of linear functions. This may present bias in situations where the distribution of data is highly nonlinear in feature space, which can result in poor model accuracy. Conversely,
neural networks may offer high performance as they are able to model the non-linear data distribution more effectively, but are less interpretable to the user. Keeping in mind the design requirements is important during this stage to know on what grounds to evaluate different models. It is commonplace to generate multiple models and to compare their performance on an unseen data set, known as the test set. The presence of a test set helps to ensure that the model has effectively modeled the distribution of data and has not overfit to points in the training set. The design criteria help to determine what metrics should be used in comparing the performance of various models. For example, if accuracy is the dominant metric, the model with the highest accuracy should be chosen. This stage, called **model validation** can also be used to determine whether specific features improved or hindered model performance (e.g., by analyzing the relative importance of features or by successively removing variables and retraining the model, known as ablation study). The last stage of the lifecycle involves the **model deployment** in the production environment. This is one of the most important stages: the model size must be reduced sufficiently to fit on the embedded device through various techniques. The model must then be compiled into a format
compatible with the end device. After all of these stages, continuous monitoring of the model in the production environment is typically necessary to ensure that it is working effectively. If, after all of these steps, the performance of the model is unsatisfactory in the production environment, the model must be updated. This can be done using new design requirements, data, or modeling techniques. 

In conclusion, TinyML can be everywhere, powering smart embedded devices in our homes and in remote locations, enabling monitoring for both industry, humans and natural settings. Today, sensors discarded the vast majority of their raw sensor data (which is a wealth of information for machine learning), becouse it is impossible to transmit all these readings to a datacenter.
